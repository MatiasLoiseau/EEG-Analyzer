```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

file_names = [
    "golpes2.dat",
    "cerrados.dat",
    "mentalimagery.dat",
    "golpes1.dat",
    "exhalar.dat",

    "inhalar.dat"

```python
column_names = ['timestamp', 'counter', 'eeg', 'attention', 'meditation', 'blinking']
```

```python
golpes2 = pd.read_csv('data/golpes2.dat', delimiter=' ', names = ['timestamp','counter','eeg','attention','meditation','blinking'])

print('Estructura de la informacion:')
print(golpes2.describe())
```

```python
data_golpes2 = golpes2.values
eeg = data_golpes2[:,2]

plt.plot(eeg,'r', label='EEG')
plt.xlabel('t');
plt.ylabel('eeg(t)');
plt.title(r'EEG Signal')
plt.ylim([-400, 400]);
plt.xlim([0,len(eeg)])
plt.show()
```

```python
# Parameters
N = len(eeg)  # Number of samples
T = 1.0 / 512.0  # Sample spacing

# Fourier transform and frequency domain plot
yf = np.fft.fft(eeg)
xf = np.linspace(0.0, 1.0/(2.0*T), N//2)

plt.plot(xf, 2.0/N * np.abs(yf[0:N//2]))
plt.grid()
plt.title(r'Signal spectrum.')
plt.ylim([0, 10])
plt.xlim([0,100])
#plt.axis((0, 60, 0, 9))
plt.show()
```
